```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
```

```python
#read in data
df=pd.read_csv(r"D:\CU\spring 2022\nlp\roku_data.csv")
```

```python
df=df.drop(columns=["Unnamed: 0"])
df.head()
```

```python
# see how many comments are in each version
df.groupby("version").count()["id"]
```

```python
# see the dist of rating
df.groupby("rating").count()["id"]
```

```python
df=df.astype({"body":str, "title":str})
```

```python
# cleaning the text, leaving only english words and numbers
def clean_txt(txt_in):
    import re
    clean_str = re.sub("[^A-Za-z0-9]+", " ", txt_in).strip().lower()
    return clean_str
df["cleaned"]=df["body"].apply(clean_txt)
```

```python
df.head()
```

```python
# count how many words are in each cleaned comments
df["cleaned_count"]=df["cleaned"].apply(lambda x: len(x.split()))
df.head()
```

```python
# see the dist of word count per comments
df.cleaned_count.describe()
```
